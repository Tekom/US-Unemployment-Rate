```python
#Looking for the distributions
fig, ax = plt.subplots(9, 3, figsize=(20,20))
fig.tight_layout(pad=3)
columns = sex_unemployment.columns
column_num = 0

for i in range(9):
    for j in range(3):
        try:
            axs = ax[i, j]
            sns.histplot(data=sex_unemployment[columns[column_num]], kde=True, ax=axs)
            column_num += 1
        except:
            pass
```

As we can see, most of the data has a exponencial distribution. some of the reasons could be:

A competitive labor market: In a competitive labor market, employers have the freedom to hire and fire workers according to the needs of their business. This can lead to fluctuating unemployment, as workers can be laid off at any time.

A globalized economy: The U.S. economy is becoming increasingly interconnected with the economies of other countries. This can make the U.S. economy more vulnerable to external shocks, such as economic crises or wars. These shocks can lead to an increase in unemployment.

Technological changes: Technology is rapidly changing the labor market. Some jobs are being automated, while others are being created. This can lead to temporary unemployment as workers adjust to the new changes in the labor market.

## iForest

```python
from sklearn.ensemble import IsolationForest

anomaly_inputs_columns = list(sex_unemployment.columns)
anomaly_inputs_columns.remove('year')
anomaly_inputs_columns.remove('date')

IF = IsolationForest(contamination=0.1, random_state=42)
```

```python
sex_unemployment = sex_unemployment.dropna()
IF.fit(sex_unemployment[anomaly_inputs_columns])
```

```python
sex_unemployment['anomaly_scores'] = IF.decision_function(sex_unemployment[anomaly_inputs_columns])
```

```python
sex_unemployment['anomaly'] = IF.predict(sex_unemployment[anomaly_inputs_columns])# -1 is considered as an outlier
```

```python
outliers = sex_unemployment[sex_unemployment['anomaly'] == -1]
outliers
```

```python
total_outliers = len(outliers)
total_outliers
```

```python
lenght_dataset = len(sex_unemployment)

print(f"Our dataset contained a {total_outliers/lenght_dataset}% of outliers")
```
